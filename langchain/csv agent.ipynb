```python
import langchain 
langchain.debug = True
agent = create_pandas_dataframe_agent(
    ChatOpenAI(temperature=0, model="gpt-3.5-turbo-0613",openai_api_key=key),
    df,
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,handle_parsing_errors=True,
)

prompt = {
    "tool_names": ["pandas"],
    "tools": {
        "pandas": {
            "df": df.to_dict()
        }
    },
    "input": "Which are the top 3 states with maximum representation in the dataset for high AQI. Enlist the top 6 counties in each of the states with maximum good days. Summarize in simple english"
}


x=agent(prompt)
```
